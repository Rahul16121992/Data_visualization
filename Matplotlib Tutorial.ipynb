# Python Matplotlib 

# Lineplot

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.title("Simple Line plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title("Simple Line plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title("Simple Line plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Adding Subplot
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)


plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)

plt.show()
```

# Bar PLot

```python
student={'Bob':87,'Matt':56,"Sam":27}

names=list(student.keys())
marks=list(student.values())

plt.bar(names,marks,color='g')
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.show()
```

```python
student={'Bob':87,'Matt':56,"Sam":27}

names=list(student.keys())
marks=list(student.values())

plt.bar(names,marks,color='g')
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

```python
student={'Bob':87,'Matt':56,"Sam":27}

names=list(student.keys())
marks=list(student.values())

plt.barh(names,marks,color='g')
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,3,0,5,7,8,9]

plt.scatter(x,a,color='r',marker='*',s=100)
plt.title("Scatter Plot")

plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,3,0,5,7,8,9]
b=[7,2,4,6,8,9,1,2,4]

plt.scatter(x,a,color='r',marker='*',s=100)
plt.scatter(x,b,color='g',marker='*',s=200)
plt.title("Scatter Plot")

plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

# Histogram

```python
data=[1,2,3,4,5,6,7,8,9,3,4,6,7,8]

plt.hist(data,color='g')
plt.show()
```

# Boxplot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data=list([one,two,three])

plt.boxplot(data)
plt.show()
```

# Violin Plot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data=list([one,two,three])

plt.violinplot(data,showmedians=True,showextrema=True,showmeans=True)
plt.show()
```

# Pie Chart

```python
fruit=["Apple","Orange","Mango","Guava"]
quantity=[67,34,100,29]

plt.pie(quantity,labels=fruit,autopct='%0.1f%%',
        colors=['yellow','blue','red','green'])

plt.show()
```

# DoughNut-Chart

```python
fruit=["Apple","Orange","Mango","Guava"]
quantity=[67,34,100,29]

plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%',
        colors=['orange','blue','red','green'])
plt.pie([1],colors=['w'],radius=1)

plt.show()
```
